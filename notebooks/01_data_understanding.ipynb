``Validate financial calculations``

```Expected Gross Amount```

````python
df["calculated_gross_price"] = (
    df["unit_price_usd"]
    * df["quantity"]
)
````

```Expected discount```


````python
df["calculated_discount_amount"] = (
    df["calculated_gross_price"]
    * df["discount_percent"]
    / 100
)
````

```Compare discount amount with expected discount amount```

````python
discount_difference = (
    df["discount_amount_usd"]
    - df["calculated_discount_amount"]
).abs()

discount_mismatches = (
    discount_difference > 0.05
).sum()

print(
    f"Discount calculation mismatches: "
    f"{discount_mismatches:,}"
)
````

```Test Possible total-price formula```

````python
df["calculated_net_price"] = (
    df["calculated_gross_price"]
    - df["discount_amount_usd"]
)
````

````python
formula_tests = pd.DataFrame({
    "net_product_price": (
        df["calculated_gross_price"]
        - df["discount_amount_usd"]
    ),

    "net_plus_tax": (
        df["calculated_gross_price"]
        - df["discount_amount_usd"]
        + df["tax_usd"]
    ),

    "net_plus_shipping": (
        df["calculated_gross_price"]
        - df["discount_amount_usd"]
        + df["shipping_cost_usd"]
    ),

    "net_plus_tax_shipping": (
        df["calculated_gross_price"]
        - df["discount_amount_usd"]
        + df["tax_usd"]
        + df["shipping_cost_usd"]
    ),
})

for formula in formula_tests.columns:
    average_difference = (
        df["total_price_usd"]
        - formula_tests[formula]
    ).abs().mean()

    print(
        f"{formula}: average difference "
        f"${average_difference:,.2f}"
    )
````

````python
profit_formula_tests = pd.DataFrame({
    "revenue_minus_cost": (
        df["total_price_usd"]
        - df["cost_usd"]
    ),

    "revenue_minus_cost_shipping": (
        df["total_price_usd"]
        - df["cost_usd"]
        - df["shipping_cost_usd"]
    ),

    "revenue_minus_cost_tax": (
        df["total_price_usd"]
        - df["cost_usd"]
        - df["tax_usd"]
    ),

    "revenue_minus_cost_tax_shipping": (
        df["total_price_usd"]
        - df["cost_usd"]
        - df["tax_usd"]
        - df["shipping_cost_usd"]
    ),
})

for formula in profit_formula_tests.columns:
    average_difference = (
        df["profit_usd"]
        - profit_formula_tests[formula]
    ).abs().mean()

    print(
        f"{formula}: average difference "
        f"${average_difference:,.2f}"
    )
````

### Observation

The dataset contains several records where:

- Order Status = Cancelled
- Payment Status = Paid
- Delivery Status = Delivered
- Return Reason = Missing

From a business perspective, this combination is inconsistent.

Possible explanations include:

- The synthetic dataset generated order and delivery statuses independently.
- The cancellation happened after delivery but no return information was recorded.
- The dataset intentionally contains noisy or inconsistent records for analytics practice.

These records will not be removed at this stage. They will be flagged during the data cleaning phase if necessary.
