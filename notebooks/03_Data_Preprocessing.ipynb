# 03. Data Preprocessing & Feature Engineering

## Objective

This notebook prepares the cleaned dataset for SQL analysis and Power BI dashboard development.

The preprocessing stage includes:

- Data cleaning
- Handling missing values
- Data type optimization
- Feature engineering
- Data validation
- Exporting the final analytics dataset

````python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/raw/ecommerce_dataset_+1m.csv")
````

````python
df["order_date"] = pd.to_datetime(df["order_date"])
df["account_creation_date"] = pd.to_datetime(df["account_creation_date"])
````

````python
missing = (
    df.isna()
      .sum()
      .to_frame("Missing")
)

missing["Percentage"] = (
    missing["Missing"] / len(df)
) * 100

missing.sort_values(
    "Missing",
    ascending=False
)
````

````python
df["return_reason"] = df["return_reason"].fillna("Not Returned")

df["coupon_code"] = df["coupon_code"].fillna("No Coupon")

df["customer_feedback"] = df["customer_feedback"].fillna("No Feedback")
````

````python
df.duplicated().sum()
````

````python
df.info(memory_usage="deep")
````

```Feature Engineering```

```Profit per Unit```

````python
df["profit_per_unit"] = (
    df["profit_usd"]
    / df["quantity"]
)

df["profit_per_unit"].head(2)
````

```Revenue per Unit```

````python
df["revenue_per_unit"] = (
    df["total_price_usd"]
    / df["quantity"]
)

df["revenue_per_unit"].head(2)
````

```Customer Age Groups```


````python
df["age_group"] = pd.cut(

    df["age"],

    bins=[18,25,35,45,55,65,100],

    labels=[
        "18-25",
        "26-35",
        "36-45",
        "46-55",
        "56-65",
        "65+"
    ]

)
````

```Order Month Name```


````python
df["month_name"] = (
    df["order_date"]
    .dt.month_name()
)
````

```Quarter```


````python
df["quarter"] = (
    df["order_date"]
    .dt.quarter
)
````

```Weekend Order```


````python
df["is_weekend"] = (
    df["order_date"]
    .dt.dayofweek >= 5
)
````

```High Value Order```

````python
df["high_value_order"] = (
    df["total_price_usd"] >= 1000
)
````
